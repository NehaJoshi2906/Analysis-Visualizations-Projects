```python
# first visualization:creating the correlation map
# we are gonna drop three columns ie:key,explicit and mode and apply pearson correlation method then we use heatmap function from seaborn 
#to get the required correlation map
#where as annot is annotation value is true which will write the data value in ecah cell 
#fmt =1g is the sting formatting tool while adding annotataions,
#vmin and vmax are the values set to anchor the colour map 
#cmap="infereno" stands for colour map where as the infereno is the colur palette used 
# then we set the title for the map and then the required labels 

corr_df =df_tracks.drop(["key","mode"],axis=1 ).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df,annot=True,fmt=".1g", vmin=1 ,vmax=1 ,center=0, cmap="inferno", linewidths =1 , linecolor="Black")
heatmap.set_title("correlation hetmap between variables ")
heatmap.set_xticklabels(heatmap.get_xticklabels()  , rotation =90)

#minus value indicates they have negative correlation ; like energy and acoustiness has high negative correlation 
#means if energy is high the acoustiness is very low where as danceability and valence have positive cotrelation.


```

```python
# we will sample aour data and take only 4% of the total data and create two regression plotes using this data
sample_df=df_tracks.sample(int(0.004*len(df_tracks)))
```

```python
print(len(sample_df))
```

```python
# we will plot the regression plots now ,we will plot it between loudness and energy as it has positive correlation as obserevd from the correlation graph
#we are using the function called regplot from the seaborn package 
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y ="loudness",x ="energy",color="c").set(title = "Loudness vs Energy")

# as we can see all the points follow the same direction and the direction of line is also in one direction gradually increasing which indicates positive corrrelation
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y ="popularity",x ="acousticness",color="c").set(title = "Loudness vs Energy")
#as we can see hte line going downward later we can see the negative correlation as popularity increase acousticness decreases
```

```python
#we are creating the distribution plots to know the distribution of songs per month 
#so we use the diplot function for creating the plot,
time_signature=df_tracks.time_signature
sns.displot(time_signature,discrete =True,aspect=2,height=5,kind="hist").set(title="no of songs acc to time ")
```

```python
# to see the duration of the songs over the time so we are gonna create the power plot
total_duration=df_tracks.duration_ms
time_signature=df_tracks.time_signature
fig_dimensions=(18,7)
fig, ax =plt.subplots(figsize=fig_dimensions)
fig=sns.barplot(x=time_signature,y=total_duration,ax =ax,errwidth = False ).set(title="Duration over the time")

```

```python
# to see duration of songs acc to different genre
plt.title("duration of songs acc to different genre")
sns.color_palette( "rocket" , as_cmap =True)
sns.barplot(y ="genre", x ="duration_ms", data =df_tracks)
plt.xlabel("duration of songs")
plt.ylabel("Genre")



```

```python
# to find genre by popularity
sns.set_style(style="darkgrid")
plt.figure(figsize=(10,6))
famous=df_tracks.sort_values("popularity"  ,ascending =False ).head(10)
sns.barplot(y="genre",x="popularity",data=famous).set(title="top 5 genre by popularity")
# we took .head(10) instead of 5 just for top 5 as there are few genre which are repetative as we have children appearing twice in our duaration map
```
