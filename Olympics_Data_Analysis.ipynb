```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns  
```

```python

regions=pd.read_csv('/content/noc_regions.csv')
```

```python
# joining two datsets using the commin column NOC
#we will be doing horizontal joint using merge function usinh how to indicate left join on means using which column
atheletes_df=atheletes.merge(regions ,how ='left' ,on ='NOC')
atheletes_df
# now we have two more colimns region and notes in our atheletes datset
```

```python
# to seee the shape of the datset
atheletes_df.shape
```

```python
# as many column names begin with capital letters but the new column names begin with small letter so to make the column name consistent 
# we will are using the rename function
#column name consistent

atheletes_df.rename(columns={'region':'Region' ,'notes':'Notes'},inplace=True)
```

```python
# to know the top 10 participating countries since the olympics started 
top_10_county =atheletes_df.Team.value_counts().sort_values(ascending=False).head(10)

top_10_county
```

```python
# we will create the bar plot for the top 10 countries who participated in olympics
plt.figure(figsize=(10,6))
plt.title("top 10 participating countries ")
sns.barplot(x =top_10_county.index , y =top_10_county, palette ='Set2');
```

```python

#to have the age distribution of participants
# so we will create histogram and we use tghe bin which states it's value starts from 10 ..will go upto 80,,with 2 as the size
plt.figure(figsize=(12,6))
plt.title("Age distribution of atheletes")
plt.xlabel("Age")
plt.ylabel("Number of participants")
plt.hist(atheletes_df.Age, bins =np.arange(10,80,2),color='pink',edgecolor='white')
```

```python
#to extract the sport details for winter sports
# we are using the unique function to get the all the possible sports without repeating the same values 
winter_sports =atheletes_df[atheletes_df.Season=='Winter'].Sport.unique()
winter_sports
```

```python
#Summer olympics sports details
Summer_sports =atheletes_df[atheletes_df.Season=='Summer'].Sport.unique()
Summer_sports
```

```python
# to know the male and female participants counts 
gender_Counts=atheletes_df.Sex.value_counts()
gender_Counts

```

```python
# plot pie chart for male and female participants
plt.figure(figsize=(10,6))
plt.title("Gender Distribution")
plt.pie(gender_Counts,labels=gender_Counts.index,autopct = '%1.1f%%' , startangle=180, shadow=True)

```

```python
# total no of female atheletes participated in olympics
female_part=atheletes_df[(atheletes_df.Sex=='F') & (atheletes_df.Season =='Summer')][ ['Sex','Year']]
female_part=female_part.groupby('Year').count().reset_index()
female_part.tail()
# tail to get the recent data 
```

```python
female_participation=atheletes_df[(atheletes_df.Sex=='F') & (atheletes_df.Season =='Summer')]
```

```python
# another way to filter the data accordin to count functiion in sns in graph format
sns.set(style="darkgrid")
plt.figure(figsize=(20,10))
sns.countplot(x='Year',data=female_participation ,palette="Spectral")
plt.title("Women participation")

```

```python
part = female_participation.groupby('Year')['Sex'].value_counts
plt.figure(figsize=(20,10))
female_part.plot()
plt.title("Women participation over time")

```

```python
#gold medals atheletes
gold_medals=atheletes_df[(atheletes_df.Medal=='Gold')]
gold_medals
```

```python
#to take values that are diffrent from Nan
gold_medals=gold_medals[np.isfinite(gold_medals['Age'])]
gold_medals
```

```python
# to get participants who won medal who are aged more then 60
gold_medals['ID'][gold_medals['Age'] > 60].count()
```

```python
# for which sport have these 6 gold medals won 
sporting_event=gold_medals['Sport'][gold_medals['Age'] > 60]
sporting_event
```

```python
#plot for sporting event
plt.figure(figsize=(10,6))
plt.tight_layout()
sns.countplot(sporting_event)
plt.title("sporting event scored gold medals for age greater then 60")

```

```python
#gold medals for each country ; to know which ciuntry has recieved more gold medals
gold_medals.Region.value_counts().reset_index(name='Medal').head(5)
```

```python
# drawing a plot to know which country recived the more no of gold medals
totalgoldmedals=gold_medals.Region.value_counts().reset_index(name='Medal').head(6)
g=sns.catplot(x="index",y="Medal",data=totalgoldmedals,
              height=5,kind="bar", palette="rocket")
g.despine(left=True)
g.set_xlabels("Top 5 counties with max gold medals")
g.set_ylabels("No of medals")
plt.title("countries wining the maximum no of gold medals")



```

```python
#RIO olympics
# looking for the recent olympics data
max_year=atheletes_df.Year.max()
max_year
```

```python
# extracting info regarding the team which won most gold medals in recent year olympics 
team_names =atheletes_df[(atheletes_df.Year==max_year) & (atheletes_df.Medal=='Gold')].Team
team_names.value_counts().head(10)
```

```python
# create the bar plot for the above results
sns.barplot(x=team_names.value_counts().head(20) ,y=team_names.value_counts().head(20).index)
plt.ylabel(None);
plt.xlabel(" countrywise medals acc to recent Olympics")
```

```python
# creating a scatter plot regardin the weight and height for male and female atheletes who won the medal
#as many atheletes did not win any medals so medal column has nan values 
not_null_medals =atheletes_df[(atheletes_df['Height'].notnull()) & (atheletes_df['Weight'].notnull())]
```
